## Tree Terminologies

A Decision tree is a flowchart-like tree structure, where each internal node denotes a test on an attribute, each branch represents an outcome of the test, and each leaf node (terminal node) holds a class label. 

A vertex (or node) of a graph is one of the objects that are connected together. The connections between the vertices are called edges or links

A connected acyclic graph is called a tree. In other words, a connected graph with no cycles is called a tree

# How to decide threshold (e.g. here 35 for Age ?)

- Here in age we assumed 35 as threshold which would give least impure nodes however algorithm will try all possible feature values as stated below & choose the one which gives least impurity for age feature & hence then age feature will be split based on that threshold.
- For categorical features, the algorithm creates a separate branch for each possible value of the feature & then calculates impurity of feature.

More pure means a better split which results in a smaller tree arriving at faster decisions.

Impurity metric measures the impurity or randomness of a dataset.


## Some impurity metrics for classification decision trees, Entropy and Gini

## <font color=blue>2 major algo's for deciding order of features is ID3 (based on entropy) and CART (based on Gini Index).


Difference between these 2 algorithms is the way you select the features & conditions under which u select the algorithm.
For ID3 you order your features based on Entropy. Based on entropy you compute Information Gain.
For CART you order your features based on Gini Index. Based on Gini Index you compute Gini gain.

- Information gain and Gini gain can be thought of as like objective functions which you want to maximize. Entropy is related to Information gain and Gini Impurity is related to Gini gain. You want to order your features in such a way that u get maximum Information gain or Gini gain (whichever criteria chosen).
- In decision tree its not about learning parameters but deciding which way u would like to split & even if u choose different/wrong feature to split it might not have much impact, its just a different decision tree you would form that is relatively larger.


- Entropy is an information theory metric. 
- In the context of Decision Trees, entropy is a measure of disorder or impurity in a node.It determines how a decision tree chooses to split data. 
- The more the decrease in the entropy, the more is the information gained. 
- Low entropy shall always be preferred over high entropy.
- Entropy of 0 means data is perfectly classified.
- The entropy of our target variable is 1, at maximum disorder, due to the even split between class label.
- Entropy always lies between 0 to 1
