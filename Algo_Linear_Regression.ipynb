## <font color=blue>Minimization of Error function (any one of mentioned above but mse preferred because easy derivative calculation)

### **<font color=blue>Result** - error minimization means we able to get best parameters (coefficients and weights).


### **<font color=blue>Application** - The best parameters result in best fit line/equation (y=mx+c)which can be used for prediction of new unseen inputs.

## <font color=blue>Measure of performance/ evaluation metric for regression models - R SQUARE (Goodness of fit measure/also called as coefficient of determination)

**<font color=blue>R square is always b/w 0 and 1, or 0 and 100%**

## Limitations of R Square

- A good model can have low R squared.
- A biased model can have a high R squared.
- Some fields of study have inherently greater amount of variations, so in those cases ur R squared values r bound to be lower. For eg in studies that try to explain human behaviour generally have R squared values less than 50%. 
- Fotunately if u have low R square values but independent variables r statistically significant u can still draw imp. conclusions.
- Statistically significant coefficents continue to represent the mean change in dependent variable given a 1 unit shift in independent variable.
- In cases where predictions needed which r generally precise, a low R square can be a show stopper.
- How high R square need to be for model to produce useful predictions that depends on precition u require & amount of variation present in ur data.
- In high R square, if chance of systemetic bias then a residual plot can give a better idea. Residual vs fit model can show this unwanted pattern.
- An unbiased model has residuals that r randomly scattered around 0. 
- Non-random residual patterns indicate bad fit despite high R squared. Always check reidual plots cz R squared alone will not tell if model is biased. This can happen when your model is mising significant independent variables, polynomial terms & interaction terms. 
- To produce random residuals, try adding terms to models when fitting a non-linear model.

### High R Squared but systematic pattern in residual plot indicating a biased model. Ideally the residuals should be randomly scattered around 0.

### Consider R Square with residual plots, other statistics and indepth knowledge of the subject area to overcome these limitations of R square.
