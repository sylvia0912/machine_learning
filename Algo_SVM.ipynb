## Constrained Optimization Problem

At the core if it is a binary classification SVM boils down to a straight line. In general finding a line, plane, hyperplane depending on dimensions of data.

SVM works very well if classes are linearly separable (linearly seperable refers to classes where you can draw a straight line and they r seperable).

An SVM CLassifier is a stright line if we take binary case as above where you have 2 features and 2 classes as shown above. The model/equation will be that of a straight line.

## Maximum margin classifier

E.g. in linear regression we optimize m and c to get best fit line y=mx+c.

- In above figure we say that red is better than green line because green line is comparatively closer to red class points compared to purple class points while red line seems more or less equidistant from both classes.
- We say red line is biased as when a new data point comes in test it may have more chances of being accurate if it belongs to purple class as lot of margin while classifier is closer to red data points and less margin.

**SVM tries to find a line which is in a way equidistant from both classes.**
- The name SVM comes from fact that how do you determine which line is equidistant from both classes. In training of SVM in given training data we look for boundary data points.
- The finding of line and boundary points happen together in optimization.

## SVM Kernel (it is a function)

The kernel takes in base features as input & it helps in creating linearly seperable features in higher dimensions. SVM is able to handle process of creating these extra features based on base features & therefore it does not require our interface. But we still need to choose the type of kernel we want to use to create these new features in higher dimensional space.rbf kernel (guassian kernel) is very useful when data is not linear.
