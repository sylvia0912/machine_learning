## ML Introduction

**Traditional way** - human written computer program for filtering out spam messages. First humans analyse historical data & decide on rules and then rules fed into a traditional human written computer program to decide whether an incoming email is spam.

**An alternate way/ML way** - instead of human written computer program to solve this problem we can have an algorithm which would autmatically read historical data & generate a series of rules. This is what ML does. AI/ML is not magic, it is just mathematics. Also called as mathematical modelling.

### Definition of ML


- ML is a sub-field of computer science that gives computers the ability to learn without being explicitly programmed. 
- Computers have the ability to learn rules without we being have to program them.

## ML tasks under Unsupervised Category

1. Clustering
2. Recommendation Engines

Clustering is finding similar patterns in a given dataset e.g. grouping segments of customers who r similar on certain set of attributes like demography, buying behaviour etc. 
Recommendation Engine discover the rules that suggest services, products & information to users. Both r unsupervised & hence do not require labels.

### Supervised



An algorithm is trained in such a way that column to be predicted by algorithm is present in the training data.

- Supervised learning in ML can be described in terms of **FUNCTION APPROXIMATION**, that is approximating the unknown underlying function that maps examples of inputs to outputs in order to make predictions on new data. Function approximation is solved using function optimization.

- It can be challenging as there is often a limited number of examples from which we can approximate the function, and the structure of the function that is being approximated is often nonlinear, noisy, and may even contain contradictions.

### Unsupervised

### ML Components


Developing machine learning applications can be viewed as consisting of three components-

1. a representation of data
2. an evaluation function (objective functions or loss functions or cost function or error function) - generally every algorithm has same cost function (mae,mse,rmse etc) but the parameters w.r.t which they are optimized are different in different algorithms talking about supervised algorithms.
3. an optimization method to estimate the parameter of the machine learning model. This optimization method can vary from algorithm to algorithm e.g. Adam, Gradient descent, BGFS etc

## ML vs Statistics

### 1. Similarities

- Machine learning is built upon a statistical framework. This should be overtly obvious since ML involves data, and data has to be described using a statistical framework.
- ML draws upon a large number of other fields of mathematics and computer science.
- **ML theory** from fields like mathematics & statistics
- **ML algorithms** from fields like optimization, matrix algebra, calculus
- **ML implementations** from computer science & engineering concepts (e.g. kernel tricks, feature hashing)

### 2. Differences

- We are trying to minimize our empirical risk in ML, this is not an inherent feature of statistics.
