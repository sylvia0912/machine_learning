- Logistic regression is a generalized linear model. And it uses the same basic formula of linear regression.
- Logistic regression uses the Sigmoid function to predict the output class label for a given input.

Sigmoid function clusters values towards 0 and 1 hence used in logistic regression where we need to do binary classification.
