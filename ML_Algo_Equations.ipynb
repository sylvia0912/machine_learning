## <font color = blue>$$2. Decision Tree Classifier$$

It is a rule based algorithm, no single mapping function as such b/w y and X i.e dependent and independent features rather at each step of algorithm, minimization of entropy required to reach the pure nodes (containing all samples of same class).

## <font color = blue>$$3. KNN$$

It is a distance based algorithm. If features are similar means they have similar numerical values and hence will be closer in space when calculated using distances b/w 2 points.
