# K Nearest Neighbours

### Based on feature similarity

KNN used for classification and regression tasks. (Supervised, has labels) while K-means is a Clustering algorithm (Unsupervised, no labels).

Binary or Multiclass classification can be done.

KNN classifies the new test data based on distance metrics. 

- We check for the new orange data point whether its features/properties (here height & weight) r similar to which of the classes, either cat or dog. 
- So whichever points r nearer to it in distance, in space x,y or higher dimensional plane, means they have properties most similar to test data point(orange point).
- Here if u choose K=3, since 2 cats have similar properties to test data point & 1 dog has similar properties to test data so we take the majority & assign test point to cat.
- Now if we take k=7, we have 3 cats and 4 dogs having similar properties to test data point. So we choose dog.
